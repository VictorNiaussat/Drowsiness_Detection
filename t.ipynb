```python
path, list_model = list_model_path()
```

```python
import os
m = load_model(os.path.join(path, list_model[0]), specs = dict(input_size=224, nb_couches_rentrainement=4, nb_classes=10) )
```

```python
video_path = load_video_path()[0]
```

```python
import cv2
cap=cv2.VideoCapture('Data/' + video_path)

n_frames= int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
n_fps = int(cap.get(cv2.CAP_PROP_FPS))

cap.set(cv2.CAP_PROP_POS_FRAMES, 200*n_fps*2)
ret,img=cap.read()
image = cv2.resize(img[200:, 700:-250],(224,224))
image_rescale = image/255.
Image.fromarray(image)
```

```python
classe = list(activity_map.values())
preds = m.predict(np.array([image]))[0]
print({k: v for k, v in zip(activity_map, preds)})
print(classe[np.argmax(preds)])
```
